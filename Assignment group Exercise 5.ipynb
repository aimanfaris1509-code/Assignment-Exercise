on the function written in Q2(d) to convert the following temperatures
125.70F, 275.40F, 300.00F, 37.480F, 0.00F, −180.450F to Celsius and print the results in tabular form.

```python
def celsius(f):
    return (f - 32) * 5/9


temps_f = [32, 68, 86, 104, 212]


results = list(map(celsius, temps_f))

print("Results using map():")
print("Fahrenheit\tCelsius")
print("-" * 25)

for f, c in zip(temps_f, results):
    print(f"{f}\t\t{c:.2f}")
```

4. Create a Python function that calculates and returns the total surface area and the volume of a

(a) rectangular box. The inputs are the length, width and the height of the box.


```python
def box_calculations(length, width, height):
    surface_area = 2 * (length * width + length * height * width * height)
    volume = length * width * height

    return {
        "Surface Area": surface_area,
        "Volume": volume
    }

print(box_calculations(5, 3, 2))
```

(b) closed cone. The inputs are the radius and the height of the cone.

```python
import math

def cone_calculations(radius, height):
    slant_height = math.sqrt(radius**2 + height**2)
    surface_area = math.pi * radius**2 + math.pi * radius * slant_height
    volume = (1/3) * math.pi * radius**2 * height

    return {
        "Surface Area": surface_area,
        "Volume": volume
    }

print(cone_calculations(3,4))
```

7. (a) Create a function that returns the average of a list of numbers.

```python
import math

def calculate_average(numbers):
    
    return sum(numbers) / len(numbers) if numbers else 0
```

```python
import math  

def calculate_average(numbers):
    
    return sum(numbers) / len(numbers) if numbers else 0

def calculate_std_dev(numbers):  
    if not numbers:
        return 0

    mu = calculate_average(numbers)

    squared_diffs = [(x - mu)**2 for x in numbers]

    variance = calculate_average(squared_diffs)

    return math.sqrt(variance)

data = [10, 12, 23, 23, 16, 23, 21, 16]

print(f"Average: {calculate_average(data)}")
print(f"Standard Deviation: {calculate_std_dev(data):.2f}")
```

9. Create the list of integers with the following pattern


[1, 4, 2, 4, 2, 4, ... , 2, 4, 1]


by returning it from a function. Note that the list starts and ends with 1, has odd-index elements
equal to 4 and even-indexed elements equal to 2. The function receives the length of the list as the
input. The length must by an odd number, if it is even, return an empty list.

```python
def create_list(n):  
    if n % 2 == 0:
        return []
    
    result = []
    for i in range(n):
        if i == 0 or i == n - 1:
            result.append(1)
        elif i % 2 == 0:
            result.append(2)
        else:
            result.append(4)
   
    return result
```
