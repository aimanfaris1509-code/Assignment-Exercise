1. Create an array for:

i. f(x)= x2sin(x) + e−2x for x ∈ [0,2] in steps of 0.1

ii.  g(x)=xsinh−1(x)+ln(x)for20equallyspacedpointsin[1,5]

iii.  h(x)=log4(x)forx=[2,4,8,16,...,210]

```python
import numpy as np

x1 = np.arange(0, 2.1, 0.1)
f_x = x1**2 * np.sin(x1) + np.exp(-2*x1)
print("1(a) i. f(x) results:\n", f_x)
```

ii. g(x)= x sinh−1(x) + ln(x) for 20 equally spaced points in[1,5]

```python
x2 = np.linspace(1, 5, 20)
g_x = x2 * np.arcsinh(x2) + np.log(x2)
print("\n1(a) ii. g(x) results:\n", g_x)
```

iii. h(x) = log4(x) for x = [2,4,8,16,...,2^10]

```python
x3 = 2**np.arange(1, 11)
h_x = np.log2(x3) / np.log2(4)
print("\n1(a) iii. h(x) results:\n", h_x)
```

find x·b, xA, Ab and A+bx.

```python
import numpy as np

x = np.array([-1, 3, 2])
A = np.array([[1, 4, 0],
              [3, 7, 5],
              [2, 8, -2]])
b = np.array([[4], [-1], [6]])

x_dot_b = x @b
xA = x @ A
Ab = A @ b
A_plus_bx = A + (b @ x.reshape(1, -1))

print("x * b =", x_dot_b)
print("xA =", xA)
print("Ab =\n", Ab)
print("A + bx =\n", A_plus_bx)
```

2.Referring to the quantities defined in Q1(b),

```python
import numpy as np


A = np.array([[1, 4, 0], 
              [3, 7, 5], 
              [2, 8, -2]])


column_sums = np.sum(A, axis=0)

print("3(b) Column sums of A:")
print(column_sums)
```

(c) sort the rows of A in descending order

```python
import numpy as np


A = np.array([[1, 4, 0], 
              [3, 7, 5], 
              [2, 8, -2]])


sorted_A_desc = np.sort(A, axis=1)[:, ::-1]

print("3(c) Rows of A sorted in descending order:")
print(sorted_A_desc)
```
